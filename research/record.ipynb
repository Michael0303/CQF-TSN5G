```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

# Define the paths to your scalar and vector files
# scalar_file_path = "results/UDPDL-numUEs=1-#0.sca"
# vector_file_path = "results/UDPDL-numUEs=1-#0.vec"
csv_file_path = "results/result.csv"
```

```python
# Load the scalar and vector files into pandas DataFrames
result = pd.read_csv(csv_file_path)
```

```python
pd.set_option('display.width', 180)
pd.set_option('display.max_colwidth', 100)
result.run.head()
```

```python
# Filter the DataFrame for entries that include 'byte', 'length', or 'size' in the name
packet_length_df = result[result['name'].str.contains("byte|length|size", case=False, na=False)]

# Further filter for sum of outgoing packet lengths
packet_length_sum_df = packet_length_df[packet_length_df['name'] == 'outgoingPacketLengths:sum']
```

```python
# Display the extracted data
print(packet_length_sum_df)

# Calculate the total and average packet size if possible
total_packet_length = packet_length_sum_df['value'].astype(float).sum()
total_packets = packet_length_sum_df['count'].astype(float).sum()  # If count information is available
average_packet_size = total_packet_length / total_packets if total_packets else None

print(f"Total Packet Length: {total_packet_length} bytes")
print(f"Average Packet Size: {average_packet_size} bytes")
```
